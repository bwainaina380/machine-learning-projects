```python
%matplotlib inline
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
boston = datasets.load_boston()
```

```python
X = boston.data
y = boston.target
```

```python
print(X.shape)
print(y.shape)
```

```python
# Linear regression model
l_reg = linear_model.LinearRegression()
```

```python
# Visualize the data
# We need to transpose because each instance has 13 attributes but the y hdata has only one column
# Below, we are using the first column of the attributes against the one column of the features
plt.scatter(X.T[0], y)
plt.show()
```

```python
plt.scatter(X.T[5], y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = l_reg.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
print(predictions)
```

```python
# Print R^2 and coefficient (slope), intercept
print(l_reg.score(X, y))
print(l_reg.coef_)
print(l_reg.intercept_)
```
